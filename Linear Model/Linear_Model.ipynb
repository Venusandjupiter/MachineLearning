```python
import pandas as pd

df = pd.read_csv("insurance.csv")
df.describe(),df.isnull().sum(),df.corr()
```

```python
df['sex'].replace({'male':1,'female':2},inplace=True)
df['smoker'].replace({'yes':1,'no':0},inplace=True)
df.drop(['region'],axis=1,inplace=True)
df
```

```python
df.corr()
```

```python
features = df.iloc[:,0:-1]
labels = df.iloc[:,-1]
features,labels
```

```python
import numpy
features = numpy.array(features,dtype=numpy.float32)
labels = numpy.array(labels,dtype=numpy.float32)
features,labels
```

```python
import torch
import torch.nn as nn
import torch.optim
import torch.utils.data

class container(torch.utils.data.Dataset):
    def __init__(self,features,labels):
        self.X =torch.from_numpy(features).cuda()
        self.Y =torch.from_numpy(labels).reshape(-1,1).cuda()
    def __getitem__(self,index):
        return self.X[index],self.Y[index]
    def __len__(self):
        return len(self.Y)
    
```

```python
set = container(features,labels)
```

```python
iterator = torch.utils.data.DataLoader(dataset=set,batch_size=128,shuffle=True,drop_last=True)
```

```python
next(iter(iterator))
```

```python
F1 = torch.from_numpy(features).cuda()
L1 = torch.from_numpy(labels).reshape(-1,1).cuda()
```

```python
Net = nn.Sequential(nn.Linear(5,1,device='cuda:0'))
nn.init.normal_(Net[0].weight,std=0.02)
optimzer = torch.optim.SGD(Net.parameters(),lr=0.0412/128)
l = nn.MSELoss()
epochs = 50
for epoch in range(epochs):
    for X in iterator:
        loss =l(Net(X[0]),X[1])
        optimzer.zero_grad()
        loss.backward()
        optimzer.step()
    with torch.no_grad():
        print(f'epoch:{epoch+1},loss:{l(Net(F1),L1).mean()}')
```
