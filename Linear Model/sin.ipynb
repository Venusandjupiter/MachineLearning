```python
import torch
import numpy
x = numpy.arange(0,6.28,0.01,dtype=numpy.float32)
x
```

```python
y =  numpy.array(numpy.sin(x),dtype=numpy.float32)
y
```

```python
def get_data():
    x = numpy.arange(0,6.28,0.01,dtype=numpy.float32)
    y =  numpy.array(numpy.sin(x),dtype=numpy.float32)
    return x,y
```

```python
features = torch.from_numpy(x)
labels = torch.from_numpy(y)
features,labels
```

```python
features=features.reshape(-1,1).cuda()
labels=labels.reshape(-1,1).cuda()
features,labels,features.shape
```

```python
import torch.nn as nn
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.net = nn.Sequential(
            nn.Linear(1,16),
            nn.Sigmoid(),
            nn.Linear(16,1)
        )
    def forward(self,x):
        return self.net(x)
```

```python
Loss = nn.MSELoss()
```

```python
import torch.utils.data
class Data(torch.utils.data.Dataset):
    def __init__(self,X,Y):
        self.features = X
        self.labels = Y
    def __getitem__(self, item):
        return self.features[item],self.labels[item]
    def __len__(self):
        return len(labels)
train = Data(features,labels)
data = torch.utils.data.DataLoader(dataset=train,shuffle=True,batch_size=128,drop_last=False)
next(iter(data))
```

```python
import torch.optim
epochs = 3000
model = Net().cuda()
optimizer = torch.optim.SGD(model.parameters(),lr=0.26,momentum=0.4)
for epoch in range(epochs):
    model.train()
    for x,y in data:
        x = x.cuda()
        y = y.cuda()
        optimizer.zero_grad()
        pred = model(x)
        loss = Loss(pred,y)
        loss.backward()
        optimizer.step()
    with torch.no_grad():
        print(f'epochs:{epoch+1},loss:{Loss(model(features),labels).mean()}')
```

```python
features = features.cuda()
res = model(features)
res = torch.detach(res.cpu())
features,res
```

```python
import matplotlib.pyplot as plt
x,y = get_data()
plt.plot(x,y,color='red',label='sin(x)')
plt.plot(x,res,color='blue',label='pred_sin(x)')
plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.show()
```
